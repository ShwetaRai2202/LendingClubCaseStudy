```python
#Importing the Python data  libraries
import numpy as np
import pandas as pd
#Importing the Python data  Visualization libraries
import matplotlib.pyplot as plt
import seaborn as sbn
import warnings
warnings.filterwarnings('ignore') # to ignore the warnings
```

```python
#Reading the data from csv file and loading it into dataframe
dataframe = pd.read_csv('loan.csv')
```

```python
# Checking the head of the data frame
dataframe.head()
```

```python
# checking unnecessary value in  'loan_status'
dataframe.loan_status.value_counts()
```

```python
#Remove the records where Loan status is not completed
dataframe=dataframe[dataframe['loan_status'] !='Current']
dataframe.head()
```

```python
#Deleting the unnecessary columns with all null values and zero values
dataframe=dataframe.dropna(axis=1,how='all')
dataframe.head()
```

```python
#checking the unique value in each column
dataframe.nunique().sort_values()
```

```python
#Now analysing columns and their description and their impact on Loan defaulting and remove such columns which seem to have no impact

non_impacting_columns=['desc', 'mths_since_last_delinq', 'mths_since_last_record','delinq_2yrs', 'earliest_cr_line', 'last_pymnt_amnt','inq_last_6mths', 'open_acc', 'pub_rec', 
                  'revol_bal', 'revol_util', 'total_acc', 'total_pymnt','total_pymnt_inv', 
                  'total_rec_prncp', 'total_rec_int', 'total_rec_late_fee', 'recoveries', 'collection_recovery_fee',
                  'last_pymnt_d','last_credit_pull_d','title','emp_title','member_id','url','tax_liens','acc_now_delinq', 'chargeoff_within_12_mths', 'delinq_amnt', 'initial_list_status', 'out_prncp', 'out_prncp_inv',
       'collections_12_mths_ex_med', 'policy_code', 'application_type','pymnt_plan','zip_code']

dataframe=dataframe.drop(non_impacting_columns,axis=1)
```

```python
dataframe.shape
```

```python
#checking all the colums after cleaning unnecessary columns
dataframe.columns
```

```python
#replacing the % to get the intrest rate
dataframe.int_rate=dataframe.int_rate.str.replace("%","")
dataframe["int_rate"] = pd.to_numeric(dataframe["int_rate"])
dataframe.dtypes

```

```python
#replacing the months to get the values in numbers
dataframe.term.str.replace("months","")
```

```python
#changing the datatype of Loan Amount and funded amount to float
dataframe.astype({'loan_amnt':'float','funded_amnt':'float'})
```

```python
#Changing the datatype of issue_d from object type to datetime type
dataframe['issue_d'] = pd.to_datetime(dataframe.issue_d, format='%b-%y')
dataframe['issue_d']
```

```python
dataframe.isnull().mean()
```

```python
# adding the year column to the dataframe 
dataframe["issue_yr"] = pd.DatetimeIndex(dataframe["issue_d"]).year
dataframe.head()
```

```python
#Removing the unwanted colums of installment, id, issue_yr and issue_mon
drp_columns = ['installment', 'id', 'issue_yr']
correlation_loan = dataframe.drop(columns=drp_columns)
correlation_loan
# Finding correlation matrix
correlation_matrix = correlation_loan.corr(numeric_only=True)

plt.figure(figsize=(10, 10))

# Heatmap with correlation matrix using a different colormap
sbn.heatmap(correlation_matrix, annot=True, cmap='YlGnBu')
plt.title("Correlation Among Columns")
plt.tight_layout()
plt.show()

#This shows 
#Inverse Relationships:
# - There is a negative correlation between the loan amounts requested (loan_amnt) and the incidences of public record bankruptcies (pub_rec_bankruptcies).
# -Similarly, the funded amounts (funded_amnt) and annual income exhibit negative correlations with debt-to-income ratio (dti).
#Moderate Associations:
#-The size of the loan (loan_amnt) shows moderate positive correlations with the loan duration (term).
#-The loan duration (term) also moderately correlates with the interest rate charged (int_rate).
#Strong Connections:
#-Strong positive correlations exist between the loan amounts (loan_amnt) and the actual funded amounts (funded_amnt).
#-Additionally, the funded amount from investors (funded_amnt_inv) demonstrates a robust correlation with the funded amount (funded_amnt).
```
